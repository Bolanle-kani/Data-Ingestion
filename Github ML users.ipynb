```python
keywords_users = 'Machine Learning location:Lagos'
```

```python
user_result = {"name":[],"username":[],"bio":[],"blog":[],"company":[],"email":[],"followers":[],"following":[],"id":[],"html_url":[],
                   "location":[],"hireable":[],"public_repos":[],"total_private_repos":[]}
```

```python
def search_github_users(keywords_users):
    rate_limit = g.get_rate_limit()
    rate = rate_limit.search
    if rate.remaining == 0:
        print('You have 0/{} API calls remaining. Reset time: {}'.format(rate.limit, rate.reset))
        return
    else:
        print('You have {}/{} API calls remaining'.format(rate.remaining, rate.limit))

    query = keywords_users
    results = g.search_users(query, sort='followers', order='desc')
 
    print('Found {} users'. format(results.totalCount))
 
    for user in results:
        # Storing results to a dataframe

        user_result["name"].append(user.name)
        user_result["username"].append(user.login)
        user_result["bio"].append(user.bio)
        user_result["blog"].append(user.blog)
        user_result["company"].append(user.company)
        user_result["email"].append(user.email)
        user_result["followers"].append(user.followers)
        user_result["following"].append(user.following)
        user_result["id"].append(user.id)
        user_result["html_url"].append(user.html_url)
        user_result["location"].append(user.location)
        user_result["hireable"].append(user.hireable)
        user_result["public_repos"].append(user.public_repos)
        user_result["total_private_repos"].append(user.total_private_repos)
```

```python
search_github_users(keywords_users)
data1 = pd.DataFrame(data=user_result)
data1.shape
```
